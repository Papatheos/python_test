```python
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np
```

```python
data = pd.read_csv('all_data.csv')
print(data.Country.unique())
```

```python
def life_expectancy_plot(country):
    # x and y iterate the dictionary and select the values for a specific country
    y=[data['Life expectancy at birth (years)'][i] for i in range(len(data)) if (data['Country'][i] == country)]
    x=[data['Year'][i] for i in range(len(data)) if (data['Country'][i] == country)]
    #here we plot!
    plt.plot(x,y)
    plt.title( country +' life expectancy')
    plt.xlabel('Year')
    plt.ylabel('Life expectancy')
    plt.show()
```

```python
for country in data.Country.unique():
    life_expectancy_plot(country)
```
